```python
import pandas as pd
```

```python
df = pd.read_excel("./project_grade_2024_12_29.xlsx")
df.head()
```

```python
import re


def formatGroupName(text):
    out = text.strip()
    out = re.sub(r"\s+", " ", out)
    return out


df["group_name"] = df["group_name"].fillna("")
df["group_name"] = df["group_name"].apply(formatGroupName)
df.head()
```

```python
def makeGroupKey(text):
    out = re.sub(r"\s+", "", text)
    out = out.lower()
    return out


df["group_key"] = df["group_name"].apply(makeGroupKey)
df["group_key"].value_counts()
```

```python
df.to_csv("project_grade.csv", index=False)
```
